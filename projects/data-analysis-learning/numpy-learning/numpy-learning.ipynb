### 3.ndarray的创建



-简单创建

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

-copy

```python
arr1 = np.copy(arr)  # 创建一个新的数组，内容与 arr 相同
print(arr1)
arr1[0] = 10  # 修改 arr1 的第一个元素
print("Modified arr1:", arr1)  # 输出修改后的 arr1
print("Original arr:", arr)  # 输出原始的 arr，验证其未被修改
```

-预定义形状的创建
（1）全0 （2）全1 （3）未初始化 （4）自定义 

```python
# 全0数组
arr = np.zeros((2, 3), dtype=int)  # 创建一个 2 行 3 列的全0数组
print(arr)
```

```python
# 全1数组
arr = np.ones((2, 3), dtype=int)  # 创建一个 2 行 3 列的全1数组
print(arr)
```

```python
# 未初始化数组
arr = np.empty((2, 3))  # 创建一个 2 行 3 列的未初始化数组
print(arr)  
```

```python
# 自定义值数组
arr = np.full((2, 3), 2026)  # 创建一个 2 行 3 列的数组，所有元素都是 2026
print(arr)
```

```python
# np.xxx_likes() 函数可以创建一个与给定数组形状相同的数组，但元素值根据函数不同而不同。例如：
# np.zeros_like() 创建一个与给定数组形状相同的全0数组
arr1 = np.zeros_like(arr)  # 创建一个与 arr 形状相同的全0数组
print(arr1) 
# np.ones_like() 创建一个与给定数组形状相同的全1数组
arr2 = np.ones_like(arr)  # 创建一个与 arr 形状相同的全1数组
print(arr2)
# np.empty_like() 创建一个与给定数组形状相同的未初始化数组
arr3 = np.empty_like(arr)  # 创建一个与 arr 形状相同的未初始化数组
print(arr3)     
# np.full_like() 创建一个与给定数组形状相同的数组，所有元素都设置为指定值
arr4 = np.full_like(arr, 2026)  # 创建一个与 arr 形状相同的数组，所有元素都是 2026
print(arr4)
```

-特殊矩阵

```python
# 单位矩阵
arr = np.eye(3)  # 创建一个 3x3 的单位矩阵
print(arr)
```

```python
# 对称矩阵  
arr = np.diag([1, 2, 3])  # 创建一个对角线元素为 1, 2, 3 的对称矩阵
print(arr)
```

-随机矩阵

```python
# 0-1的浮点数随机矩阵
arr = np.random.rand(2, 3)  # 创建一个 2 行 3 列的随机浮点数矩阵，元素值在 [0, 1) 范围内
print(arr)
```

```python
# 任意区间的浮点数随机矩阵
arr = np.random.uniform(10, 20, (2, 3))  # 创建一个 2 行 3 列的随机浮点数矩阵，元素值在 [10, 20) 范围内
print(arr)  
```

```python
# 任意区间的整数随机矩阵
arr = np.random.randint(10, 20, (2, 3))  # 创建一个 2 行 3 列的随机整数矩阵，元素值在 [10, 20) 范围内
print(arr)  
```

```python
# 任意区间的正态分布随机矩阵
arr =np.random.randn(2, 3)  # 创建一个 2 行 3 列的随机矩阵，元素值服从标准正态分布（均值为0，标准差为1）
print(arr)
```

```python
# 设置随机种子
np.random.seed(42)  # 设置随机种子为 42
arr = np.random.rand(2, 3)  # 创建一个 2 行 3 列的随机浮点数矩阵，元素值在 [0, 1) 范围内
print(arr)  # 每次运行时都会得到相同的随机数矩阵
```

### 5.索引与切片



#一维数组的索引与切片

```python
arr =np.random.randint(1,100, 20)
print(arr)
```

```python
print(arr[0])  # 输出第一个元素
```

### 6.矩阵的算术运算

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
b = b.T  # 转置操作，虽然对于一维数组没有实际效果，但在这里为了演示转置的概念

print(a +b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)  
```

```python
# 矩阵的点乘运算
print(a)
print(b)

print(a @ b)  # 使用 @ 运算符进行矩阵的点乘运算
```

#基本数学函数

```python
# 开平方根运算
arr = np.array([1, 4, 9, 16, 25])
print(arr)
print(np.sqrt(arr))  # 输出每个元素的平方根
```

```python
# 指数运算: e^x
arr = np.array([0,1, 2, 3, 4, 5])
print(arr)
print(np.exp(arr))  # 输出每个元素的指数值 
```

```python
# 对数运算： ln x
arr = np.array([1, np.e, np.e**2, np.e**3])  # 创建一个数组，包含 1 和 e 的幂次方
print(arr)
print(np.log(arr))  # 输出每个元素的自然对数值
```

```python
# 绝对值运算： ｜x｜
arr = np.array([-1, -2, 3, -4, 5])
print(arr)
print(np.abs(arr))  # 输出每个元素的绝对值
```

```python
# 幂函数 ： x^n
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(np.power(arr, 2))  # 输出每个元素的平方值
```

```python
# 四舍五入运算
arr = np.array([1.2, 2.5, 3.7, 4.4, 5.9])
print(arr)
print(np.round(arr))  # 输出每个元素四舍五入后的值，0.5向下取，0.51向上取
```

```python
# 上取整，下取整
arr = np.array([1.2, 2.5, 3.7, 4.4, 5.9])
print(arr)
print(np.ceil(arr))  # 输出每个元素的上取整值
print(np.floor(arr))  # 输出每个元素的下取整值
```

```python
# 检测缺失值 NAN
arr = np.array([1, 2, np.nan, 4, 5])
print(arr)
print(np.isnan(arr))  # 输出一个布尔数组，指示每个元素是否是 NaN（True 表示是 NaN，False 表示不是 NaN）
```

#统计函数

-求和、平均值、中位数、标准差、方差
-最大值、最小值
-计算分位数、累积和、累积差

```python
arr = np.random.randint(1,20,8)
print(arr)
```

```python
print(np.sum(arr))  # 输出数组元素的总和
print(np.mean(arr))  # 输出数组元素的平均值
print(np.median(arr))  # 输出数组元素的中位数   
print(np.max(arr))  # 输出数组元素的最大值
print(np.min(arr))  # 输出数组元素的最小值
```

```python
# 数组的标准差
print(np.std(arr))  # 输出数组元素的标准差
# 数组的方差
print(np.var(arr))  # 输出数组元素的方差
```

```python
# 数组的百分位数
print(np.percentile(arr, 25))  # 输出数组元素的第 25 百分位数
print(np.percentile(arr, 50))  # 输出数组元素的第 50 百分位数（即中位数）
print(np.percentile(arr, 75))  # 输出数组元素的第 75 百分位数
```

```python
#数组的累积和
print(np.cumsum(arr))  # 输出数组元素的累积和
#数组的累积积
print(np.cumprod(arr))  # 输出数组元素的累积积
```

#比较函数
大于、小于、等于；
与、或、非；
是否至少有一个True、全都是True；
自定义比较；

```python
arr = np.array([3,4,5,6,7,8])
print(arr)
```

```python
print(np.greater(arr, 5))  # 输出一个布尔数组，指示每个元素是否大于 5（True 表示大于 5，False 表示不大于 5）
print(np.less(arr, 5))  # 输出一个布尔数组，指示每个元素是否小于 5（True 表示小于 5，False 表示不小于 5）
print(np.equal(arr, 5))  # 输出一个布尔数组，指示每个元素是否等于 5（True 表示等于 5，False 表示不等于 5）
```

```python
print(np.logical_and([0,1],[1,0]))  # 输出一个布尔数组，指示每个元素是否同时满足两个条件（True 表示同时满足，False 表示不满足）
print(np.logical_or([0,1],[1,0]))  # 输出一个布尔数组，指示每个元素是否满足至少一个条件（True 表示满足至少一个条件，False 表示不满足任何条件）
print(np.logical_not([0,1]))  # 输出一个布尔数组，指示每个元素的逻辑非（True 表示原来是 False，False 表示原来是 True）
```

```python
print(np.any([0, 1, 0]))  # 输出 True，因为至少有一个元素为 True
print(np.all([0, 0, 1]))  # 输出 False，因为没有任何元素为 True
```

```python
# np.where(条件，满足条件,不满足条件) 函数可以根据条件返回满足条件的元素的索引或元素值。例如：
print(np.where((arr > 5),1,0))  # 输出满足条件 arr > 5 的元素的索引
# np.select(条件列表, 满足条件的结果列表, 不满足条件的结果) 函数可以根据多个条件返回满足条件的元素的索引或元素值。例如：
conditions = [arr < 5, arr == 5, arr > 5]
choices = [0, 1, 2]
print(np.select(conditions, choices, default=-1))  # 输出满足条件的元素的索引或元素值，如果没有满足条件的元素，则输出 -1
```
