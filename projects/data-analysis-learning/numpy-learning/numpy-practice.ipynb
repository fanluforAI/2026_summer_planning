```python
"""
题目1: 温度数据分析
某城市一周的最高气温为 [28,30,29,31,32,30,29]
- 计算平均气温、最高气温、最低气温
- 找出气温高于30的天数
"""
```

```python
import numpy as np
```

```python
arr = np.array([28,30,29,31,32,30,29])
```

```python
print(np.mean(arr))  # 输出平均气温
print(np.max(arr))   # 输出最高气温
print(np.min(arr))   # 输出最低气温
```

```python
print(np.sum(arr > 30))  # 输出气温高于30的天数
```

```python
"""

题目 2:学生成绩统计
某班级 5 名学生的数学成绩为[85,90,78,92, 88]
- 计算成绩的平均分、中位数和标准差
- 将成绩转换为10分制(假设满分为 10)

"""
```

```python
arr = np.array([85,90,78,92, 88])
print("5student scores:", arr)
```

```python
print("average score:", np.mean(arr))  # 输出平均分
print("median score:",np.median(arr))  # 输出中位数
print("standard deviation:", np.std(arr))  # 输出标准差
```

```python
print("10-point scores:", arr/10)  # 将成绩转换为10分制
```

```python
"""
题目 3:矩阵运算
给定矩阵A=[[1,2],[3,4]]和B=[[5,6],[7,8]]
- 计算A+B和A*B(逐元素乘法)。
- 计算A和B的矩阵乘法(点积)。

"""
```

```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
print("A:", arr1)
print("B:", arr2)
```

```python
print("A+B:", arr1 + arr2)  # 逐元素加法
print("A*B (element-wise):", arr1 * arr2)  # 逐元素乘法
print("A dot B:", np.dot(arr1, arr2))  # 矩阵乘法(点积)
```

```python
"""
题目 4:随机数据生成
生成一个(3,4)的随机整数数组，范围[0,10)。
- 计算每列的最大值和每行的最小值。
- 将数组中的所有奇数替换为 -1。
"""
```

```python
arr = np.random.randint(0, 10, (3,4))
print("Random array:\n", arr)
```

```python
for i in range(0,3):
    print(np.max(arr[i,:]))
for j in range(0,4):
    print(np.min(arr[:,j]))
```

```python
print(np.where(arr % 2 == 1, -1, arr))  # 将数组中的所有奇数替换为 -1
```

```python
"""
题目 6：布尔索引
题目：生成一个 (5, 5) 的随机数组，范围 [0, 20)。  
- 找出数组中大于 10 的元素  
- 将所有大于 10 的元素替换为 0
"""
```

```python
arr =np.random.randint(0, 20, (5,5))
print("Random array:\n", arr)
```

```python
print(np.greater(arr, 10))  # 找出数组中大于 10 的元素
print(arr[arr > 10])  # 输出大于 10 的元素
print(np.where(arr > 10, 0, arr))  # 将所有大于 10 的元素替换为 0
```

```python
"""
题目 7：统计函数应用
题目：某公司 6 个月的销售额（万元）为 [120, 135, 110, 125, 130, 140]。  
- 计算销售额的总和、均值和方差。  
- 找出销售额最高的月份和最低的月份。
"""
```

```python
arr = np.array([120, 135, 110, 125, 130, 140])
print("Total sales:", np.sum(arr))  # 输出销售额的总和
print("Average sales:", np.mean(arr))  # 输出销售额的均值
print("Sales variance:", np.var(arr))  # 输出销售额的方差
```

```python
print("销售额最高的月份: ", np.argmax(arr) + 1)  # 输出销售额最高的月份
print("销售额最低的月份: ", np.argmin(arr) + 1)  # 输出销售额最低的月份
```

```python
"""
题目 8：数组拼接
题目：给定 A = [1, 2, 3] 和 B = [4, 5, 6]。  
•水平拼接为 [1, 2, 3, 4, 5, 6]。  
•垂直拼接为 [[1, 2, 3], [4, 5, 6]]。  
"""
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

```python
print("水平拼接:", np.concatenate((arr1, arr2)))  # 水平拼接
print("垂直拼接:\n", np.vstack((arr1, arr2)))  # 垂直拼接
```
