## series 的创建

```python
import pandas as pd
import numpy as np
s = pd.Series([1, 2, 3, 4, 5])
print(s)
```

```python
# 自定义创建
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)
```

```python
# 是用字典格式创建
s = pd.Series({'a': 10, 'b': 20, 'c': 30}, name='cost')
print(s)
```

## Series的属性

```python
print("index: ", s.index)
print("values: ", s.values)
print("dtype: ", s.dtype)
print("size: ", s.size)
print("shape: ", s.shape)
print("ndim: ", s.ndim)
print("name: ", s.name)
```

```python
print("隐式索引，通过位置访问: ", s.iloc[0])  # 隐式索引，通过位置访问
print("显式索引，通过标签访问: ", s.loc['a'])  # 显式索引，通过标签访问
print("显式索引，通过标签访问:\n ", s.loc['a':'c'])  # 显式索引，通过标签访问
```

```python
print(s[s<30])  # 条件筛选
```

## Series案例

```python
""" 
创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。计算平均分、最高分、最低分，并找出高于平均分的学生人数。
"""
```

```python
arr = np.random.randint(50, 101, size=10)
print("学生成绩: ", arr)
```

```python
s = pd.Series(arr,name = "math_scores")
print(s)
print("平均分: ", s.mean())
print("最高分: ", s.max())
print("最低分: ", s.min())
print("找出高于平均分的学生人数: " ,s[s>s.mean()].count())
```

```python
""" 
给定某城市一周每天的最高温度Series，完成以下任务：
找出温度超过30度的天数
计算平均温度
将温度从高到低排序
找出温度变化最大的两天
"""
```

```python
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33], 
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
```

```python
print("温度超过30度的天数", temperatures[temperatures>30].count())
print("平均气温：", temperatures.mean())
temperatures.sort_values(ascending=False)
```

```python
# ⭐⭐⭐找出温度变化最大的两天
t3 = temperatures.diff().abs().sort_values(ascending=False)
print(t3)
print(t3.index[0:2])
```

```python
""" 
股票价格分析
给定某股票连续10个交易日的收盘价Series
计算每日收益率（当日收盘价/前日收盘价 - 1)
找出收益率最高和最低的日期
计算波动率（收益率的标准差）
"""
```

```python
import pandas as pd
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                  index=pd.date_range('2023-01-01', periods=10))
```

```python
for i in range(prices.count()-1):
    daily_return[i] = prices.iloc[i+1]/prices.iloc[i] - 1
    print(daily_return)
    print(daily_return)

```

```python
""""
销售数据分析
某产品过去12个月的销售量Series：
-计算季度平均销量（每3个月为一个季度）
-找出销量最高的月份
-计算月环比增长率
-找出连续增长超过2个月的月份
"""
```

```python
import numpy as np
import pandas as pd
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                 index=pd.date_range('2022-01-01', periods=12,freq="MS"))
print(sales)
```

```python
print("计算季度平均销量（每3个月为一个季度）:\n ", sales.resample("QS").mean())
print("________________")
print("找出销量最高的月份: " , sales.idxmax())
print("________________")

print("计算月环比增长率: ", sales.pct_change())
 
```

```python
b = sales.pct_change() > 0
b[b.rolling(3).sum() == 3].index.tolist()
```

```python
"""
商店每小时销售额：
-按天重采样计算每日销售额
-计算每日营业时间与非营业时间的销售额比例
-找出销售额最高的3个小时
"""
```

```python
np.random.seed(42)
hour_sales = pd.Series(np.random.randint(0,100,24), index=pd.date_range("2026-1-1", freq='h',periods=24))
hour_sales
```

```python
day_sales = hour_sales.resample("D").sum()
print(day_sales)
```

```python
hour_sales.sort_values(ascending=False).iloc[0:3]
```
